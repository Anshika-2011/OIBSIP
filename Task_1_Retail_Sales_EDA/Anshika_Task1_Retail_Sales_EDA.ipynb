```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("RETAIL_SALES-checkpoint.csv")
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df["Month"] = df["Date"].dt.month_name()
```

```python
monthly_sales = df.groupby("Month")["Total Amount"].sum()
monthly_sales
```

```python
month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly_sales = df.groupby("Month")["Total Amount"].sum().reindex(month_order)

monthly_sales.plot(kind="bar", figsize=(8,5))
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
gender_sales = df.groupby("Gender")["Total Amount"].sum()
gender_sales
```

```python
gender_sales.plot(kind="bar", figsize=(5,4))
plt.title("Gender-wise Sales")
plt.xlabel("Gender")
plt.ylabel("Total Sales")
plt.show()
```

```python
age_sales = df.groupby("Age")["Total Amount"].sum()
age_sales.head()
```

```python
age_sales.plot(kind="line", figsize=(10,5))
plt.title("Age-wise Sales")
plt.xlabel("Age")
plt.ylabel("Total Sales")
plt.show()
```

```python
category_sales = df.groupby("Product Category")["Total Amount"].sum()
category_sales
```

```python
category_sales.plot(kind="bar", figsize=(6,4))
plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
quantity = df.groupby("Product Category")["Quantity"].sum()
quantity
```

```python
quantity.plot(kind="bar", figsize=(6,4))
plt.title("Quantity Sold by Category")
plt.xlabel("Product Category")
plt.ylabel("Quantity")
plt.show()
```
